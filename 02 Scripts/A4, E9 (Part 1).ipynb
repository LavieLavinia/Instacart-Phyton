# 04. Data Wrangling

```python
# rename column surnam
df_cust.rename(columns = {'Surnam' : 'Last Name'}, inplace = True)
```

```python
# rename column STATE
df_cust.rename(columns = {'STATE' : 'State'}, inplace = True)
```

```python
# rename column fam_status
df_cust.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
```

```python
# changing data type for user_id in customer data
df_cust['user_id'] = df_cust['user_id'].astype('str')
```

```python
# changing data type for user_id in ords_prods data
ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('str')
```

# 05. Data Consistency Check

```python
# check for missing values
df_cust.isnull().sum()
```

```python
# create a subset for these missing values
df_nan = df_cust[df_cust['First Name'].isnull() == True]
df_nan

# The missing first names is about 5.5% from total data, I decided to leave the missing values as it is.
```

```python
# check for any mixed-typed columns
for col in df_cust.columns.tolist(): 
    weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis =1) 
    if len(df_cust[weird]) > 0: 
        print(col)
```

```python
# fix the mixed-type data issue for 'First Name' column
df_cust['First Name'] =df_cust['First Name'].astype('str')
```

```python
# recheck for any mixed-typed columns
for col in df_cust.columns.tolist(): 
    weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis =1) 
    if len(df_cust[weird]) > 0: 
        print(col)
```

```python
# recheck for missing values
df_cust.isna().sum()

# by changing the 'first name' column to a string datatype, the empty values (empty strings) are no longer considered as null values.
```

```python
# check for duplications
df_cust[df_cust.duplicated()]

# There is no duplicates found
```

# 06. Combining Data

```python
# merging the customer data and ords_prods data using "user_id" as a key and the argument how = 'left'
df_merged = ords_prods_merge.merge(df_cust, on = 'user_id', how = 'left')
```
